````python
# Initialize client
client = Groq()

# Generate speech
speech_file_path = Path("speech.wav")
response = client.audio.speech.create(
    model="playai-tts-arabic",
    voice="Amira-PlayAI",
    response_format="wav",
    input="ÿßÿ≤ŸäŸÉ Ÿäÿß ÿ¨ÿØÿπ ÿπÿßŸÖŸÑ ÿßŸäŸá ŸÅŸäŸÜŸÉ ŸÖŸÜ ÿ≤ŸÖÿßŸÜ ÿßŸÜÿß ŸáŸäÿ´ŸÖ ŸÖŸÜ ŸÖÿµÿ± ŸàÿßŸÜÿ™ ŸÖŸäŸÜ ÿ®ÿ∏ÿ®ÿ∑ÿü ÿßŸÇÿØÿ± ÿßÿ≥ÿßÿπÿØŸÉ ÿßÿ≤ÿßŸäÿü",
)

# Save the audio file
with open(speech_file_path, "wb") as f:
    f.write(response.read())

# Display audio in Jupyter
Audio(filename="speech.wav")
````
